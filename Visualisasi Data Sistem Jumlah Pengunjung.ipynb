```python
#read data
df = pd.read_csv("/content/gdrive/MyDrive/SEMESTER 6/Big Data/bigpro/database.csv")
print(df.shape)
df 
```

```python
df = df.drop(columns=df.columns[0])
```

```python
print(df.shape)
df
```

```python
df.info()
```

```python
# Group data by date and count the number of visitors for each day
grouped_data = df.groupby('end_time')['total_people_inside'].sum()

# Create a bar plot of the data
grouped_data.plot(kind='bar')

# Set plot title and axis labels
plt.title('Jumlah Pengunjung Terbanyak per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Pengunjung')

# Show the plot
plt.show()
```

```python
plt.clf()
grouped_data = df.groupby('end_time') ['total_people_inside'].sum().sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Urutan Jumlah Pengunjung Terbanyak', loc='center',pad=30, fontsize=15, color='black')
plt.xlabel('Tanggal', fontsize=15)
plt.ylabel('Jumlah Pengunjung', fontsize=15)
plt.ylim(ymin=0)
plt.show()
```
